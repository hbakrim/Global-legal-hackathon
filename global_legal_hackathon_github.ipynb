```python
df = pd.read_csv('dataFile.csv',encoding = 'utf8',sep=';')
print(df.shape)
print(list(df))
df.head(10)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn import preprocessing

le1 = preprocessing.LabelEncoder()
enc1 = OneHotEncoder()
le2 = preprocessing.LabelEncoder()
enc2 = OneHotEncoder()
```

```python
listLabel = list(le1.fit_transform(df['distric']))
listLabel = [ [i]for i in listLabel]
DistrictEnco = enc1.fit_transform(listLabel).todense()
```

```python
listLabel = list(le2.fit_transform(df['categorie de réparation']))
listLabel = [ [i]for i in listLabel]
RepEnco = enc2.fit_transform(listLabel).todense()
```

```python
VecBasic = np.concatenate((DistrictEnco, RepEnco), axis=1)
```

```python
newDfX = pd.DataFrame(VecBasic)
newDfX.head()
```

```python
newDfX['Acceptation Travaux']= [ int(i=='oui') for i in df['Acceptation Travaux'] ]
newDfX['Rapport expert']= [ int(i=='oui') for i in df['Rapport expert'] ]
newDfX['Contre expertise']= [ int(i=='oui') for i in df['Contre expertise'] ]
newDfX['défenseur recidiviste']= [ int(i=='oui') for i in df['défenseur recidiviste'] ]
newDfX['type Habitation']= [ int(i=='individuel') for i in df['type Habitation'] ]

newDfX['Nombre de Partie']= df['Nombre de Partie']
newDfX['Valeur Maison']= df['Valeur Maison']
newDfX['Coût Reperation']= df['Coût Reperation']
newDfX['Année Expérience Expert']= df['Année Expérience Expert']
newDfX['durée du préjudice']= df['durée du préjudice']
newDfX['estimation du Prejudice']= df['estimation du Prejudice']

newDfY1 = pd.DataFrame()
newDfY2 = pd.DataFrame()


newDfY1['Accorder']= [ int(i=='oui') for i in df['Accorder'] ]
newDfY2['reparation']= df['reparation']
newDfY2['temps Attente Audiance']= df['temps Attente Audiance']


newDfX.head()
```

```python
X = np.array(newDfX)
Y1 = np.array(newDfY1)
Y2 = np.array(newDfY2)
print(X.shape)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=4, random_state=0)

clf.fit(X, Y1)
clf.predict(X[:35,:])
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(max_depth=4, random_state=0)

regr.fit(X, Y2)
(regr.predict(X)).astype(int)
```
